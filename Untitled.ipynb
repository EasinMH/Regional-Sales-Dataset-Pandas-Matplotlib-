```python
import pandas as pd 
import numpy as np 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sheet = pd.read_excel(
    r'C:\Users\User\Desktop\Project\Regional Sales Dataset.xlsx',
    sheet_name=None
)
```

```python
df_sales = sheet ['Sales Orders']
df_customers = sheet['Customers']
df_products = sheet['Products']
df_regions = sheet['Regions']
df_states = sheet['State Regions']
df_budgets = sheet['2017 Budgets']
```

```python
df_states.columns = df_states.iloc[0] 
df_states = df_states.drop(index=0)    
df_states.reset_index(drop=True, inplace=True) 
```

```python
df_sales.isnull().sum()
```

```python
df_sales.info()
```

```python
#Merge with customers 

df = df_sales.merge(
    df_customers,
    how = 'left',
    left_on = 'Customer Name Index',
    right_on = 'Customer Index'
)
```

```python
#Merge with Products 

df = df.merge(
    df_products,
    how = 'left',
    left_on = 'Product Description Index',
    right_on = 'Index'
)
```

```python
#Merge with Regions 

df = df.merge(
    df_regions,
    how = 'left',
    left_on = 'Delivery Region Index',
    right_on = 'id'
)
```

```python
#Merge with State Regions 

df = df.merge(
    df_states,
    how = 'left',
    left_on = 'state_code',
    right_on = 'State Code'
)
```

```python
#Merge with State Regions 

df = df.merge(
    df_budgets,
    how = 'left',
    left_on = 'Product Name',
    right_on = 'Product Name'
)
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns.values
```

```python
imp_cols = ['ordernumber', 'orderdate','channel'
             ,'order quantity', 'unit price',
             'line total', 'total unit cost',
             'customer names', 'product name', 'county',
             'state_code', 'latitude', 'longitude',
             'state','region', '2017 budgets']
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df = df [imp_cols]
```

```python
df.loc[df['orderdate'].dt.year != 2017, '2017 budgets'] = pd.NA
```

```python
df.isnull().sum()
```

```python
df_2017 = df.loc[df['orderdate'].dt.year == 2017]
```

```python
df['totalCost'] = df['order quantity'] * df['total unit cost']
```

```python
df['profit'] = df['line total'] - df['totalCost']
df['profit margin%'] = df['profit'] / df['line total'] *100
```

```python
df.rename(columns={
    'ordernumber': 'order_number',
    'orderdate': 'order_date',
    'order quantity': 'order_quantity',
    'unit price': 'unit_price',
    'line total': 'revenue',
    'total unit cost': 'total_unit_cost',
    'customer names': 'customer_names',
    'product name': 'product_name',
    'county': 'county',
    'state_code': 'state_code',
    'latitude': 'latitude',
    'longitude': 'longitude',
    'state': 'state',
    'region': 'region',
    '2017 budgets': 'budgets',
    'totalCost' : 'total_cost',
    'profit margin%' : 'profit_margin_pct'
}, inplace=True)

```

```python
df.head(5)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])


df['year_month'] = df['order_date'].dt.to_period('M')


monthly_sales = df.groupby('year_month')['revenue'].sum().reset_index()


monthly_sales['year_month'] = monthly_sales['year_month'].astype(str)


plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['year_month'], monthly_sales['revenue'], marker='o', linestyle='-', color='b')
plt.title('Overall Monthly Sales Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()


plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])


df = df[df['order_date'].dt.year.isin([2014, 2015, 2017, 2018])]


df = df[~((df['order_date'].dt.year == 2018) & (df['order_date'].dt.month.isin([1, 2])))]


df['month'] = df['order_date'].dt.month_name().str[:3]  # e.g., 'Jan', 'Feb'


monthly_sales = df.groupby('month')['revenue'].sum().reset_index()


month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales['month'] = pd.Categorical(monthly_sales['month'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('month')


plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['month'], monthly_sales['revenue'], marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Revenue (All Years)', fontsize=14)
plt.xlabel('Months For All Years', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()


plt.show()
```

```python
product_revenue = df.groupby('product_name')['revenue'].sum().reset_index()


top_10_products = product_revenue.sort_values(by='revenue', ascending=False).head(10)


plt.figure(figsize=(10, 5))
plt.bar(top_10_products['product_name'], top_10_products['revenue'], color='skyblue')
plt.title('Top 10 Products by Revenue', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])


product_revenue = df.groupby('product_name')['revenue'].sum().reset_index()


bottom_10_products = product_revenue.sort_values(by='revenue', ascending=True).head(10)


plt.figure(figsize=(12, 6))
bars = plt.bar(bottom_10_products['product_name'], bottom_10_products['revenue'], color='lightcoral')


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:,.2f}', 
             ha='center', va='bottom', fontsize=10)

plt.title('Bottom 10 Products by Revenue', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.show()
```

```python
channel_revenue = df.groupby('channel')['revenue'].sum().reset_index()


plt.figure(figsize=(5, 5))
plt.pie(channel_revenue['revenue'], labels=channel_revenue['channel'], 
        autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Revenue Distribution by Channel', fontsize=14)
plt.axis('equal')  


plt.show()
```

```python
state_revenue = df.groupby('state')['revenue'].sum().reset_index()


top_10_states = state_revenue.sort_values(by='revenue', ascending=False).head(10)


plt.figure(figsize=(14, 7))
bars = plt.bar(top_10_states['state'], top_10_states['revenue'], color='skyblue')


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:,.2f}', 
             ha='center', va='bottom', fontsize=10)

plt.title('Top 10 States by Revenue', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
